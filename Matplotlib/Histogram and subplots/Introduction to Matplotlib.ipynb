# Matplotlib Example Workflow 

```python
# import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline 
import matplotlib.pyplot as plt

# prepare data
x=[1,2,3,4]
y=[11,22,33,44]

# Setup plot 
fig,ax=plt.subplots(figsize=(10,10))

# plot the data
ax.plot(x,y)

# customize the data
ax.set(title="Simple Plot",xlabel="x-axis",ylabel="y-axis")

# show the data
plt.show()

# Save & Show (you save the whole figure)
fig.savefig("images/sample_plot.png")
```

## Making figures wuth Numpy Arrays 

we want 
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
# Create some data
x=np.linspace(0,10,100)
x[:10]
```

```python
# plot the data and create a line plot 
fig,ax=plt.subplots()
ax.plot(x,x**2)
```

```python
plt.show()
```

```python
# Use same data to plot scatter plot
fig,ax=plt.subplots()
ax.scatter(x,np.exp(x))
plt.show()
```

```python
# Another Scatter plots
fig,ax=plt.subplots()
ax.scatter(x,np.sin(x))
```

```python
plt.show()
```

```python
# make a plot form dictionary
nut_butter_prices={"Almond butter":10,
                   "Peanut butter":8,
                   "Cashew butter":12}
fig,ax=plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Store",ylabel="Prices ($)");
plt.show()
```

```python
fig,ax=plt.subplots()
ax.barh(nut_butter_prices.keys(),nut_butter_prices.values())
plt.show()
```

```python
# make some data for histogram
x=np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(x)
plt.show()
```

## Two options for subplots

```python
# Subplots option 1
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))

# plot to each different axis
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax4.hist(np.random.randn(1000))
plt.show()
```
